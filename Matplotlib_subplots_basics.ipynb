```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import subplots #i have generaly imported pyplot as plt, but lets check this out
```

```python
rng = np.random.default_rng(1303)
```

Graphics with matplotlib

```python
fig, ax = subplots(figsize=(8,8))
x = rng.standard_normal(100)
y = rng.standard_normal(100)
ax.plot(x,y); #this will plot a line, wich in this case is probably not what we want
```

```python
#the last code ploted a line, wich in this case is probably not what we want
fig, ax = subplots(figsize=(8,8))
ax.plot(x,y,'o'); #'o' asks for a scatter plot, we can also use ax.scatter(x,y)
```

```python
# we can use set_xlabel() to label our axis
fig, ax = subplots(figsize=(8,8))
ax.scatter(x,y)
ax.set_xlabel('x-axis')
ax.set_ylabel('i-axis')
ax.set_title('Plot of X vs Y');
```

```python
#we can modify the figure to alter the plot
fig.set_size_inches(12,3)
fig
```

```python
#it is also posibble to have multiple plots in one figure, by giving subplots additional arguments
fig, axes = subplots(nrows = 2,
                     ncols = 3,
                     figsize = (15,5))
```

```python
#we can populate each plot individually
axes[0,1].plot(x,y,'o')
axes[1,2].scatter(x,y,marker = '+')
fig
```

```python
#we can save the image with savefig('Name.png/pdf/jpg', dpi = n)
fig.savefig('Figure.png', dpi=400);
```

```python
# we can keep modifying the figure step by step
axes[0,1].set_xlim([-1,1])
fig.savefig('Figure_updated.jpg')
fig
```

```python
# ax.contour() allows us to represent 3D data
fig, ax = subplots(figsize = (8,8))
x = np.linspace(-np.pi, np.pi,50) #np.linespace(a, b, n) creates a vector with n numbers starting at 'a' and ending at 'b' 
y = x
f = np.multiply.outer(np.cos(y), 1/(1+x**2))
ax.contour(x,y,f);
```

```python
fig, ax=subplots(figsize=(8,8))
ax.contour(x,y,f,levels=45);
```

```python
# with ax.imshow we can create a heatmap
fig, ax=subplots(figsize=(8,8))
ax.imshow(f);
```
